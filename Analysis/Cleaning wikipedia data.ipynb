```python
import re
import pandas as pd
import clean_txt as clean


certanity = []
sentence = []

# Reading wikipedia file with sentences and assesed certanity
with open('wiki_data.txt', encoding='utf-8') as file:
    for line in file:
        certanity.append(re.search('<sentence certainty="(.*)" id', line).group(1))
        sentence.append(re.search('id=".*">(.*)</sentence>', line).group(1))
```

```python
wiki_data = pd.DataFrame([sentence, certanity]).T
wiki_data.columns = ['sentence', 'certanity']

# LAzy code to clean wiki sentences
wiki_data['sentence'] = wiki_data['sentence'].apply(lambda x: clean.to_lowercase(x))
wiki_data['sentence'] = wiki_data['sentence'].apply(lambda x: clean.remove_stopwords(x, []))
wiki_data['sentence'] = wiki_data['sentence'].apply(lambda x: clean.tokenize_text(x))
wiki_data['sentence'] = wiki_data['sentence'].apply(lambda x: clean.remove_nonunicode(x))
wiki_data['sentence'] = wiki_data['sentence'].apply(lambda x: clean.remove_nummeric(x))
wiki_data['sentence'] = wiki_data['sentence'].apply(lambda x: clean.lemmatize_verbs(x))
wiki_data['sentence'] = wiki_data['sentence'].apply(lambda x: ' '.join(x))
```

```python
wiki_data.to_csv('wiki_data_clean.csv', index=False)
```
